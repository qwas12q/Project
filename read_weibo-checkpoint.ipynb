```python
import jieba
import jieba.analyse
#
jieba.analyse.set_stop_words('.txt')
# l = jieba.cut(data[""].iloc[1])
# type(l)
```

```python
# 
segments = []
for content in weibo_data[""]:
    #TextRank 
    words = jieba.analyse.textrank(content, topK=20,withWeight=False,allowPOS=('ns', 'n', 'vn', 'v'))
    splitedStr = ''
    for word in words:
        # 
        segments.append({'word':word, 'count':1})
        splitedStr += word + ' '
# df
dfSg = pd.DataFrame(segments)

# 
dfWord = dfSg.groupby('word')['count'].sum()
```

```python
#csv
dfWord.to_csv('keywords.csv',encoding='utf-8')
```
